```python
cococola_new = cococola_data
```

```python
cococola_new['quarter_list']=0
for i in range(len(cococola_new)):
    p=cococola_new['Quarter'][i]
    cococola_new['quarter_list'][i]=p[1:2]
```

```python
cococola_new['year_list']=0
for i in range(len(cococola_new)):
    p=cococola_new['Quarter'][i]
    cococola_new['year_list'][i]=p[3:5]
```

```python
cococola_new
```

```python
df_dummies = pd.DataFrame(pd.get_dummies(cococola_new['quarter_list']))
```

```python
cococola_df = pd.concat([cococola_new,df_dummies],axis=1)
cococola_df.head()
```

```python
cococola_df['quarter_square'] = cococola_df['quarter_list'].apply(lambda x:x**2)
```

```python
from numpy import log

cococola_df['log_sales'] = cococola_df['Sales'].apply(lambda x:log(x))
```

```python
cococola_df = cococola_df.rename(columns={1:'q1',2:'q2',3:'q3',4:'q4'})
cococola_df.head()
```

```python
cococola_df.shape
```

```python
train_model=cococola_df.head(30)
test_model=cococola_df.tail(12)
```

### Model Based Forecasting

### Linear Model

```python
import statsmodels.formula.api as smf
from sklearn.metrics import mean_absolute_percentage_error
```

```python
linear_model = smf.ols('Sales~quarter_list',data=train_model).fit()
linear_pred = linear_model.predict(test_model['quarter_list'])

linear_rms = mean_absolute_percentage_error(test_model['Sales'],linear_pred)*100
linear_rms
```

### Exponential Model

```python
exp_model = smf.ols('log_sales~quarter_list',data=train_model).fit()
exp_pred = exp_model.predict(test_model['quarter_list'])

exp_rms = mean_absolute_percentage_error(test_model['log_sales'],exp_pred)*100
exp_rms
```

### Quadratic Model

```python
quad_model = smf.ols('Sales~quarter_list+quarter_square',data=train_model).fit()
quad_pred = quad_model.predict(test_model[['quarter_list','quarter_square']])

quad_rms = mean_absolute_percentage_error(test_model['Sales'],quad_pred)*100
quad_rms
```

### Additional Seasonality Model

```python
add_sea_model = smf.ols('Sales~q1+q2+q3+q4',data=train_model).fit()
add_pred = add_sea_model.predict(test_model[['q1','q2','q3','q4']])

add_rms = mean_absolute_percentage_error(test_model['Sales'],add_pred)*100
add_rms
```

### Additional Seasonality with Quadratic Model

```python
add_quad_model = smf.ols('Sales~quarter_list+quarter_square+q1+q2+q3+q4',data=train_model).fit()
add_quad_pred = add_quad_model.predict(test_model[['quarter_list','quarter_square','q1','q2','q3','q4']])

add_quad_rms = mean_absolute_percentage_error(test_model['Sales'],add_quad_pred)*100
add_quad_rms
```

### Multiplicative Seasonality Model

```python
mul_sea_model = smf.ols('log_sales~q1+q2+q3+q4',data=train_model).fit()
mul_sea_pred = mul_sea_model.predict(test_model[['q1','q2','q3','q4']])

mul_sea_rms = mean_absolute_percentage_error(test_model['log_sales'],mul_sea_pred)*100
mul_sea_rms
```

### Multiplicative with additional seasonality Model

```python
mul_add_model = smf.ols('Sales~quarter_list+q1+q2+q3+q4',data=train_model).fit()
mul_add_pred = mul_add_model.predict(test_model[['quarter_list','q1','q2','q3','q4']])

mul_add_rms = mean_absolute_percentage_error(test_model['Sales'],mul_add_pred)*100
mul_add_rms
```

```python
train_data=cococola_data.head(30)
test_data=cococola_data.tail(12)
```

### Simple Exponential Smoothing Model

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing,Holt,ExponentialSmoothing

ses_model = SimpleExpSmoothing(train_data["Sales"]).fit(smoothing_level=0.2)
pred_ses = ses_model.predict(start = test_data.index[0],end = test_data.index[-1])

ses_rms=mean_absolute_percentage_error(pred_ses,test_data['Sales'])*100 
ses_rms
```

### Holt Model

```python
hw_model = Holt(train_data["Sales"]).fit(smoothing_level=0.8, smoothing_slope=0.2)
pred_hw = hw_model.predict(start = test_data.index[0],end = test_data.index[-1])
hw_rms=mean_absolute_percentage_error(pred_hw,test_data['Sales'])*100 
hw_rms
```

### Holts winter exponential smoothing with multiplicative seasonality and additive trend

```python
hwe_model_mul_add = ExponentialSmoothing(train_data["Sales"],seasonal="mul",trend="add",seasonal_periods=12).fit() 
pred_hwe_mul_add = hwe_model_mul_add.predict(start = test_data.index[0],end = test_data.index[-1])

hw_ma_rms=mean_absolute_percentage_error(pred_hwe_mul_add,test_data['Sales'])*100 
hw_ma_rms
```

### Holts winter exponential smoothing with additive seasonality and additive trend

```python
hwe_model_add_add = ExponentialSmoothing(train_data["Sales"],seasonal="add",trend="add",seasonal_periods=12).fit() #add the trend to the model
pred_hwe_add_add = hwe_model_add_add.predict(start = test_data.index[0],end = test_data.index[-1])

hw_aa_rms=mean_absolute_percentage_error(pred_hwe_add_add,test_data['Sales'])*100 
hw_aa_rms
```

### ARMA Model

```python
from statsmodels.tsa.arima_model import ARMA,ARIMA

ARMAmodel = ARMA(train_data['Sales'],order=(1,1))
ARMAmodel_fit = ARMAmodel.fit()

ARMA_pred = ARMAmodel_fit.predict(0,11)
ARMA_pred

arma_rms=mean_absolute_percentage_error(ARMA_pred,test_data['Sales'])*100 
arma_rms
```

```python
ARIMAmodel = ARIMA(train_data['Sales'],order=(1,1,2))
ARIMAmodel_fit = ARIMAmodel.fit()

ARIMA_pred = ARIMAmodel_fit.predict(1,12,typ='levels')
ARIMA_pred

arima_rms = mean_absolute_percentage_error(ARIMA_pred,test_data['Sales'])*100
arima_rms
```

### Converting non stationary data to stationary data to improve ARIMA model

```python
rolLmean = cococola_data['Sales'].rolling(12).mean() # 12 entries
rolLstd = cococola_data['Sales'].rolling(12).std()

plt.figure(figsize=(16,7))
fig = plt.figure(1)

#Plot rolling statistics:
orig = plt.plot(cococola_data['Sales'], color='blue',label='Original')
mean = plt.plot(rolLmean, color='red', label='Rolling Mean')
std = plt.plot(rolLstd, color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

#### Log_transform

```python
plt.figure(figsize=(16,7))
fig = plt.figure(1)

import numpy as np
ts_log = np.log(cococola_data['Sales'])#to transform to stationary from non-stationary
plt.plot(ts_log)
```

### Differencing

```python
plt.figure(figsize=(16,7))
fig = plt.figure(1)
ts_log_diff = ts_log - ts_log.shift() # I will shift the time series by 1 and subtract from the original TS.
plt.plot(ts_log_diff)

#Determing rolling statistics
rolLmean = ts_log_diff.rolling(12).mean()
rolLstd = ts_log_diff.rolling(12).std()




#Plot rolling statistics:
orig = plt.plot(ts_log_diff, color='blue',label='Original')
mean = plt.plot(rolLmean, color='red', label='Rolling Mean')
std = plt.plot(rolLstd, color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

### ACF and PACF Plot

```python
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(ts_log_diff, nlags=12)
lag_pacf = pacf(ts_log_diff, nlags=12)
```

```python
import statsmodels.api as sm
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ts_log_diff.dropna(),lags=12,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ts_log_diff.dropna(),lags=12,ax=ax2)
```

```python
ts_log_diff = ts_log_diff[~ts_log_diff.isnull()]
```

```python
plt.figure(figsize=(12,8))
ts_log_diff.dropna(inplace=True)
model = ARIMA(ts_log_diff, order=(4,1,2))  
results_ARIMA = model.fit()  
plt.plot(ts_log_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
```
