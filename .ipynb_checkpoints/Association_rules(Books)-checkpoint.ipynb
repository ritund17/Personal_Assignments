```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
freq_itemsets = apriori(df=books,min_support=0.05,use_colnames=True)
freq_itemsets
```

```python
best_associates = association_rules(df=freq_itemsets,min_threshold=0.7,metric='confidence')
best_associates
```

```python
supports=[0.1,0.2,0.3,0.4]
thersholds=[0.2,0.3,0.4,0.5]

for i in supports:
    frequent_itemsets=apriori(df=books,min_support=i,use_colnames=True)
    print('support is ',i)
    print('frequent items is: ',frequent_itemsets)
    print('')
    for threshold in thersholds:
        rules=association_rules(frequent_itemsets,min_threshold=threshold)
        print('thersholds is: ',threshold)
        rules=rules.sort_values('support',ascending=False)
        print(rules)
```
