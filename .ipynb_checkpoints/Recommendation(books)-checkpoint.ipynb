### providing UBCF

```python
cosine_pivot= pd.pivot_table(data = books,values='Book.Rating',index='User.ID',columns='Book.Title',).fillna(0)
cosine_pivot
```

```python
cosine_pivot.index = books['User.ID'].unique()
cosine_pivot
```

```python
cosine_pivot.values
```

```python
from sklearn.metrics import pairwise_distances
cosine_ubcf = 1-pairwise_distances(X=cosine_pivot.values,metric='cosine')
cosine_ubcf
```

```python
cosine_ubcf         = pd.DataFrame(data=cosine_ubcf)
cosine_ubcf.index   = books['User.ID'].unique()
cosine_ubcf.columns = books['User.ID'].unique()
cosine_ubcf.round(2)
```

```python
first_50_users = cosine_ubcf.iloc[:50,:50]
first_50_users
```

```python
import numpy as np
np.fill_diagonal(a = first_50_users.to_numpy(), val = 0)
```

```python
first_50_users = cosine_ubcf.iloc[:50,:50]
first_50_users.round(2)
```
