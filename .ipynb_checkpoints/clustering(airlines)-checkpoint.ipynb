```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
airlines_data = pd.read_excel('EastWestAirlines.xlsx',1)
airlines_data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(airlines_data)
```

```python
columns = airlines_data.columns
```

```python
scaled_df = pd.DataFrame(data=scaled_data,columns=columns)
scaled_df
```

```python
dendrogram = sch.dendrogram(sch.linkage(scaled_df, method='complete'))
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'complete')
hc
```

```python
y_hc = hc.fit_predict(scaled_df)
y_hc
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
airlines_data1 = airlines_data.copy()
airlines_data1['clusters'] = Clusters
```

```python
airlines_data1
```
