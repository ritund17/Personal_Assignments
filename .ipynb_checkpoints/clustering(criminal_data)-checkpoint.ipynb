```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

import warnings
warnings.filterwarnings('ignore')
```

```python
crime_data = pd.read_csv('crime_data.csv')
crime_data
```

```python
crime_data.rename( columns={'Unnamed: 0':'states'}, inplace=True )
```

```python
del crime_data['states']
```

```python
crime_data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaled_crime_data = scaler.fit_transform(crime_data)
```

```python
columns = crime_data.columns
columns
```

```python
scaled_df = pd.DataFrame(data=scaled_crime_data,columns=columns)
scaled_df
```

```python
dendrogram = sch.dendrogram(sch.linkage(scaled_df, method='single'))
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'single')
hc
```

```python
y_hc = hc.fit_predict(scaled_df)
y_hc
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
crime_data1 = crime_data.copy()
```

```python
crime_data1['clusters'] = Clusters
```

```python
crime_data1
```

```python
crime_data1[crime_data1['clusters'] == 0]
```

```python
crime_data1[crime_data1['clusters'] == 1]
```

```python
crime_data1[crime_data1['clusters'] == 2]
```

```python
crime_data1[crime_data1['clusters'] == 3]
```

## Kmeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans_cluster = KMeans(n_clusters=4)
kmeans_cluster.fit(scaled_df)
pred = kmeans_cluster.predict(scaled_df)
```

## DBScan clustering

```python
from sklearn.cluster import DBSCAN
dbs = DBSCAN(eps=0.3, min_samples=3)
dbs.fit(scaled_df)
```

```python
dbs_clusters = dbs.labels_
dbs_clusters
```

```python
crime_data3 = crime_data.copy()
```

```python
crime_data3['dbs_clusters'] = dbs_clusters
```
